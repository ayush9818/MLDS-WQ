# Multivariate Gaussian Models

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

### Univariate Gaussian

The probability density function (PDF) of the univariate Gaussian distribution is given by:
$$f(x | \mu, \sigma^2) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2 \sigma^2}\right)$$

Where:
- $x$ is the random variable.
- $\mu$ is the mean of the distribution.
- $\sigma^2$ is the variance of the distribution.


```python
# Univariate Gaussian
mean_uni = 0
std_uni = 1
x_uni = np.linspace(-5, 5, 1000)
pdf_uni = (1 / (std_uni * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_uni - mean_uni) / std_uni) ** 2)

# Plotting
plt.figure(figsize=(8, 5))

# Univariate Gaussian
plt.plot(x_uni, pdf_uni, label='Univariate Gaussian', color='blue')
plt.title('Univariate Gaussian Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

### Multivariate Gaussian

- The multivariate Gaussian distribution describes the joint probability distribution of multiple random variables.
- It is characterized by a mean vector $μ$ and a covariance matrix $Σ$.
- Each dimension of the distribution may be correlated with other dimensions due to the covariance matrix.
- The distribution can model complex data distributions in multidimensional space.

The probability density function (PDF) of the multivariate Gaussian distribution is given by:

$$f(\mathbf{x} | \mathbf{\mu}, \mathbf{\Sigma}) = \frac{1}{(2 \pi)^{n/2} |\mathbf{\Sigma}|^{1/2}} \exp\left(-\frac{1}{2} (\mathbf{x} - \mathbf{\mu})^T \mathbf{\Sigma}^{-1} (\mathbf{x} - \mathbf{\mu})\right) $$

Where:
- ${x}$ is a vector of random variables (multivariate).
- ${\mu}$ is the mean vector of the distribution.
- ${\Sigma}$ is the covariance matrix of the distribution.
- $n$ is the number of dimensions (features) in the multivariate distribution.
- $|{\Sigma}|$ denotes the determinant of the covariance matrix.


```python
# Multivariate Gaussian
mean_multi = np.array([0, 0])
cov_multi = np.array([[1, 0.5], [0.5, 1]])
x_multi, y_multi = np.mgrid[-5:5:.01, -5:5:.01]
pos_multi = np.dstack((x_multi, y_multi))
pdf_multi = multivariate_normal(mean_multi, cov_multi).pdf(pos_multi)

# Plotting
plt.figure(figsize=(8, 5))
plt.contour(x_multi, y_multi, pdf_multi, colors='red')
plt.title('Multivariate Gaussian Distribution')
plt.xlabel('x1')
plt.ylabel('x2')

plt.tight_layout()
plt.show()
```
