## Path Finding: Shortest Path

### Problem: 
Given a set of cities and the distances between them, find the shortest path between two cities.

### Solution: 
We can model this problem as a graph, where each city is a node and the distances between them are the weights on the edges. We can then use NetworkX to find the shortest path between two cities.

In this example, we will define the distances between the cities as a dictionary, where each key is a tuple representing a pair of cities, and the value is the distance between them. We then create a graph and add the edges with their weights. We draw the graph to visualize it, and then use the shortest_path function to find the shortest path between two cities, taking into account the weights on the edges.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Define the distances between the cities
distances = {
    ('A', 'B'): 10,
    ('A', 'C'): 15,
    ('B', 'C'): 5,
    ('B', 'D'): 12,
    ('C', 'D'): 8,
    ('C', 'E'): 6,
    ('D', 'E'): 3
}

# Create a graph and add the edges with their weights
g = nx.Graph()
for edge, weight in distances.items():
    g.add_edge(edge[0], edge[1], weight=weight)

# Draw the graph
pos = nx.spring_layout(g)
nx.draw_networkx_nodes(g, pos, node_size=500)
nx.draw_networkx_labels(g, pos)
nx.draw_networkx_edges(g, pos)
nx.draw_networkx_edge_labels(g, pos, edge_labels=distances)
plt.axis('off')
plt.show()

# Find the shortest path between two cities
path = nx.shortest_path(g, 'A', 'E', weight='weight')
print('Shortest path:', path)
```

```python
#average_shortest_path_length
print(nx.average_shortest_path_length(g))
```

## Airports

### Problem: 
Given a set of airports and their flight connections, find the shortest path and flight time between two airports.

### Solution: 
We can model this problem as a graph, where the airports are nodes and the flight connections are edges with weights representing the flight time. We can then use NetworkX to find the shortest path and flight time between two airports.

In this example, we define the airports and their flight connections as a dictionary, where each key is an airport and the value is another dictionary representing the flight connections and the corresponding flight times. We then create a graph and add the nodes and edges based on the flight connections. We draw the graph to visualize it, and then use the shortest_path and shortest_path_length functions to calculate the shortest path and flight time between two airports

```python
# Define the airports and their flight connections
airports = {
    'LAX': {'SFO': 1.5, 'JFK': 6, 'DFW': 3},
    'SFO': {'LAX': 1.5, 'JFK': 6, 'SEA': 1.5},
    'JFK': {'LAX': 6, 'SFO': 6, 'SEA': 5, 'DFW': 3},
    'SEA': {'SFO': 1.5, 'JFK': 5},
    'DFW': {'LAX': 3, 'JFK': 3}
}

# Create a graph and add the nodes and edges
G = nx.Graph()
G.add_nodes_from(airports.keys())
for u in airports:
    for v in airports[u]:
        G.add_edge(u, v, weight=airports[u][v])

# Draw the graph
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=500, node_color='r')
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): airports[u][v] for u, v in G.edges()})
plt.axis('off')
plt.show()
```

```python
# Calculate the shortest path and flight time between two airports
path = nx.shortest_path(G, source='LAX', target='JFK', weight='weight')
time = nx.shortest_path_length(G, source='LAX', target='JFK', weight='weight')
print('Shortest path:', path)
print('Flight time:', time, 'hours')
```

```python
pagerank = nx.pagerank(G)
print(pagerank)
```
